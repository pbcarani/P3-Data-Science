```python
dados=ibov.join(itsa, lsuffix='_ibov', rsuffix='_itsa')
```

```python
x=dados['Date']
y=[dados['Adj Close_ibov'], dados['Adj Close_itsa']]
plt.figure(1, figsize=(16,10))
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, y[0], 'r')
ax2.plot(x, y[1], 'b')
plt.title('ITSA [2018 - 2019]', fontsize=16)
plt.xlabel('Dias', fontsize=14)
plt.ylabel('Preco', fontsize=14)
plt.show();
```

```python
plt.scatter(y[1],y[0])
```

```python
train, test = train_test_split(dados, test_size=0.20)
```

```python
X_train = np.array(train['Adj Close_itsa']).reshape(-1, 1)
y_train = train['Adj Close_ibov']
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_train)
```

```python
plt.figure(1, figsize=(16,10))
plt.title('Regresso Linear  | Treinameto' )
plt.scatter(X_train, y_train, label='Pontuao')
plt.plot(X_train, y_pred, color='r', label='Regresso')
plt.xlabel('Preo')
plt.ylabel('Pontuao')
plt.legend()
plt.show()
```

```python
X_test = np.array(test['Adj Close_itsa']).reshape(-1, 1)
y_test = test['Adj Close_ibov']
y_pred = model.predict(X_test)
```

```python
plt.figure(1, figsize=(16,10))
plt.title('Regresso Linear  |  Teste')
plt.plot(X_test,y_pred , color='r', label='Regresso')
plt.scatter(X_test, y_test, edgecolor='w', label='Pontos')
plt.xlabel('Dias')
plt.ylabel('Preo')

plt.show()
```

```python
from sklearn.metrics import explained_variance_score
explained_variance_score(y_test, y_pred)
```
